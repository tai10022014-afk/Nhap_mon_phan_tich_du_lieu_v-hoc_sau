```
import pandas as pd

url = "https://raw.githubusercontent.com/AnhNguyenVLU/Data-Analytics-and-Deep-Learning/refs/heads/main/TAI%20LIEU%20THAM%20KHAO%20LAB%20VA%20DATA/DuLieu/dulieuxettuyendaihoc.csv"
df = pd.read_csv(url)
print(df.head(10))
print(df.tail(10))
```

```
print(df.dtypes)
print(df.describe())
```

```
print(df['DT'].isnull().sum())
print(df['DT'].value_counts(dropna=False))
df['DT'] = df['DT'].fillna(0)
```

```
print(df['T1'].isnull().sum())
mean_T1 = df['T1'].mean()
df['T1'] = df['T1'].fillna(mean_T1)
```

```
diem_cols = ['T1','L1','H1','S1','V1','X1','D1','N1',
             'T2','L2','H2','S2','V2','X2','D2','N2',
             'T6','L6','H6','S6','V6','X6','D6','N6',
             'DH1','DH2','DH3']
for col in diem_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mean())
```

```
df['TBM1'] = (df['T1']+df['L1']+df['H1']+df['S1']+
              df['V1']+df['X1']+df['D1']+df['N1']) / 8
df['TBM2'] = (df['T2']+df['L2']+df['H2']+df['S2']+
              df['V2']+df['X2']+df['D2']+df['N2']) / 8
df['TBM3'] = (df['T6']+df['L6']+df['H6']+df['S6']+
              df['V6']+df['X6']+df['D6']+df['N6']) / 8

print(df[['TBM1','TBM2','TBM3']].head())
```

```
def xep_loai(tbm):
    if tbm < 5.0:   return 'Y'
    elif tbm < 6.5: return 'TB'
    elif tbm < 8.0: return 'K'
    elif tbm < 9.0: return 'G'
    else:           return 'XS'

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
print(df[['TBM1','XL1','TBM2','XL2','TBM3','XL3']].head())
```

```
def minmax_4(series):
    mn, mx = series.min(), series.max()
    return ((series - mn) / (mx - mn)) * 4

df['US_TBM1'] = minmax_4(df['TBM1'])
df['US_TBM2'] = minmax_4(df['TBM2'])
df['US_TBM3'] = minmax_4(df['TBM3'])
print(df[['US_TBM1','US_TBM2','US_TBM3']].head())
```

```
def ket_qua(row):
    kt = row['KT']
    d1, d2, d3 = row['DH1'], row['DH2'], row['DH3']
    if kt == 'A':
        diem = (d1*2 + d2 + d3) / 4
    elif kt == 'B':
        diem = (d1 + d2*2 + d3) / 4
    else:
        diem = (d1 + d2 + d3) / 3
    return 1 if diem >= 5.0 else 0

df['KQXT'] = df.apply(ket_qua, axis=1)
print(df['KQXT'].value_counts())
```
