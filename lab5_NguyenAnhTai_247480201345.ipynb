```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# ============================================================
# PHẦN 1: KHỞI TẠO VÀ ĐỌC DỮ LIỆU
# ============================================================

df = pd.read_excel("ecommerce_sales_data.xlsx")

print("=== 7 DÒNG ĐẦU ===")
print(df.head(7))

print("\n=== 5 DÒNG CUỐI ===")
print(df.tail(5))

print("\n=== THÔNG TIN CẤU TRÚC ===")
print(df.info())

print("\n=== THỐNG KÊ MÔ TẢ ===")
print(df.describe())

```

```python
# ============================================================
# PHẦN 2: KHÁM PHÁ VÀ LÀM SẠCH DỮ LIỆU
# ============================================================

print("\n=== GIÁ TRỊ BỊ THIẾU ===")
print(df.isnull().sum())

df['Discount'] = df['Discount'].fillna(0)
df['CustomerID'] = df['CustomerID'].fillna('GUEST')

print("\n=== KIỂM TRA SAU KHI XỬ LÝ ===")
print(df.isnull().sum())

df['Date'] = pd.to_datetime(df['Date'])
print("\n=== KIỂU DỮ LIỆU SAU CHUYỂN ĐỔI ===")
print(df.dtypes)

print("\n=== SỐ DÒNG TRÙNG LẶP ===")
print(df.duplicated().sum())
df = df.drop_duplicates()
print("Sau khi xóa:", df.duplicated().sum(), "dòng trùng")

```

```python
# ============================================================
# PHẦN 3: TRÍCH XUẤT VÀ BIẾN ĐỔI DỮ LIỆU
# ============================================================

df['Revenue'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.day_name()

def price_segment(price):
    if price >= 150:
        return 'High'
    elif price >= 50:
        return 'Medium'
    else:
        return 'Low'

df['Price_Segment'] = df['UnitPrice'].apply(price_segment)

print("\n=== DỮ LIỆU SAU FEATURE ENGINEERING ===")
print(df[['Revenue', 'Year', 'Month', 'DayOfWeek', 'Price_Segment']].head(10))

```

```python
# ============================================================
# PHẦN 4: LỌC VÀ TRUY VẤN DỮ LIỆU
# ============================================================

print("\n=== ĐƠN HÀNG KHU VỰC NORTH VÀ REVENUE > 300 ===")
north_high = df[(df['Region'] == 'North') & (df['Revenue'] > 300)]
print(north_high)

print("\n=== SỐ ĐƠN HÀNG ELECTRONICS KHÔNG GIẢM GIÁ ===")
elec_no_disc = df[(df['Product_Category'] == 'Electronics') & (df['Discount'] == 0)]
print(len(elec_no_disc))

print("\n=== ĐƠN HÀNG THÁNG 3 ĐẾN THÁNG 6 ===")
mar_jun = df[(df['Month'] >= 3) & (df['Month'] <= 6)]
print(mar_jun)

print("\n=== TOP 10 ĐƠN HÀNG DOANH THU CAO NHẤT ===")
top10 = df.nlargest(10, 'Revenue')[['OrderID', 'Product_Name', 'Revenue']]
print(top10)

print("\n=== ĐƠN HÀNG SỐ LƯỢNG NHIỀU NHẤT ===")
max_qty = df.loc[df['Quantity'].idxmax()]
print(max_qty)

```

```python
# ============================================================
# PHẦN 5: PHÂN TÍCH, GOM NHÓM VÀ TỔNG HỢP
# ============================================================

print("\n=== TỔNG DOANH THU THEO KHU VỰC ===")
revenue_by_region = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(revenue_by_region)
print("Khu vực doanh thu cao nhất:", revenue_by_region.idxmax())

print("\n=== TRUNG BÌNH QUANTITY VÀ UNITPRICE THEO DANH MỤC ===")
avg_by_cat = df.groupby('Product_Category')[['Quantity', 'UnitPrice']].mean()
print(avg_by_cat)

print("\n=== SỐ ĐƠN HÀNG THEO THÁNG ===")
orders_by_month = df.groupby('Month').size().sort_values(ascending=False)
print(orders_by_month)
print("Tháng nhiều đơn nhất:", orders_by_month.idxmax())

print("\n=== TOP 3 KHÁCH HÀNG CHI TIÊU CAO NHẤT ===")
top3_cust = (df[df['CustomerID'] != 'GUEST']
             .groupby('CustomerID')['Revenue']
             .sum()
             .nlargest(3))
print(top3_cust)

```

```python
# ============================================================
# PHẦN 6: TRỰC QUAN HÓA DỮ LIỆU
# ============================================================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Biểu đồ 1: Tổng doanh thu theo danh mục sản phẩm
revenue_by_cat = df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)
axes[0].bar(revenue_by_cat.index, revenue_by_cat.values, color=sns.color_palette("Set2"))
axes[0].set_title('Tổng Doanh Thu Theo Danh Mục Sản Phẩm', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Danh Mục Sản Phẩm', fontsize=11)
axes[0].set_ylabel('Tổng Doanh Thu (VNĐ)', fontsize=11)
axes[0].tick_params(axis='x', rotation=15)
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))
for i, v in enumerate(revenue_by_cat.values):
    axes[0].text(i, v + 100, f'{v:,.0f}', ha='center', fontsize=8)

# Biểu đồ 2: Tổng doanh thu theo tháng
revenue_by_month = df.groupby('Month')['Revenue'].sum()
axes[1].plot(revenue_by_month.index, revenue_by_month.values,
             marker='o', color='steelblue', linewidth=2, markersize=7)
axes[1].fill_between(revenue_by_month.index, revenue_by_month.values, alpha=0.2, color='steelblue')
axes[1].set_title('Biến Động Tổng Doanh Thu Theo Tháng', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Tháng', fontsize=11)
axes[1].set_ylabel('Tổng Doanh Thu (VNĐ)', fontsize=11)
axes[1].set_xticks(revenue_by_month.index)
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))
for x, y in zip(revenue_by_month.index, revenue_by_month.values):
    axes[1].text(x, y + 50, f'{y:,.0f}', ha='center', fontsize=7)

plt.tight_layout()
plt.savefig('ecommerce_charts.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nĐã lưu biểu đồ: ecommerce_charts.png")
```
