```python
import pandas as pd

# Vấn đề 1: Thêm header
column_names = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
df = pd.read_csv("patient_heart_rate.csv", names=column_names, on_bad_lines='skip')
print(df.head())
```

```python
# Vấn đề 2: Tách cột Name thành Firstname và Lastname
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
print(df)
```

```python
# Vấn đề 3: Chuẩn hóa cột Weight về kg
weight = df['Weight']

for i in range(0, len(weight)):
    x = str(weight[i])
    if "lbs" in x[-3:]:
        x = x[:-3]
        float_x = float(x)
        y = int(float_x / 2.2)
        y = str(y) + "kgs"
        weight[i] = y

print(df)
```

```python
# Vấn đề 4: Xóa dòng rỗng hoàn toàn
df.dropna(how="all", inplace=True)
print(df)
```

```python
# Vấn đề 5: Xóa dòng trùng lặp
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(df)
```

```python
# Vấn đề 6: Xóa ký tự non-ASCII
df.Firstname.replace(r'[^\x00-\x7F]+', '', regex=True, inplace=True)
df.Lastname.replace(r'[^\x00-\x7F]+', '', regex=True, inplace=True)
print(df)
```

```python
# Vấn đề 7: Xử lý Missing values cho Age và Weight
# Xóa dòng nào thiếu cả Age lẫn Weight
df = df[~(df['Age'].isna() & df['Weight'].isna())]

# Điền Age bằng mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Điền Weight bằng mode
df['Weight'].fillna(df['Weight'].mode()[0], inplace=True)

print(df)
```

```python
# Vấn đề 8: Tách cột m0006 thành PulseRate, Sex, Time
df = pd.melt(df, id_vars=['Id','Age','Weight','Firstname','Lastname'],
             value_name='PulseRate', var_name='sex_and_time').sort_values(['Id','Age','Weight','Firstname','Lastname'])

tmp_df = df['sex_and_time'].str.extract(r'(\b)(\d+)(\d{2})', expand=True)
tmp_df.columns = ['Sex', 'hours_lower', 'hours_upper']
tmp_df['Time'] = tmp_df['hours_lower'] + "-" + tmp_df['hours_upper']

df = pd.concat([df, tmp_df], axis=1)
df = df.drop(['sex_and_time','hours_lower','hours_upper'], axis=1)
df = df.dropna()
print(df)
```

```python
# Câu 11: Xử lý missing values cho cột huyết áp (PulseRate)
df['PulseRate'] = pd.to_numeric(df['PulseRate'], errors='coerce')

for person, group in df.groupby(['Firstname', 'Lastname']):
    idx = group.index
    for i in idx:
        if pd.isna(df.loc[i, 'PulseRate']):
            pos = list(idx).index(i)

            # 1. Trung bình liền trước và liền sau
            prev = df.loc[idx[pos-1], 'PulseRate'] if pos > 0 else None
            nxt = df.loc[idx[pos+1], 'PulseRate'] if pos < len(idx)-1 else None
            if prev and nxt:
                df.loc[i, 'PulseRate'] = (prev + nxt) / 2
                continue

            # 2. Trung bình 2 giá trị liền trước
            if pos >= 2:
                p1 = df.loc[idx[pos-1], 'PulseRate']
                p2 = df.loc[idx[pos-2], 'PulseRate']
                if not pd.isna(p1) and not pd.isna(p2):
                    df.loc[i, 'PulseRate'] = (p1 + p2) / 2
                    continue

            # 3. Trung bình 2 giá trị liền sau
            if pos <= len(idx)-3:
                n1 = df.loc[idx[pos+1], 'PulseRate']
                n2 = df.loc[idx[pos+2], 'PulseRate']
                if not pd.isna(n1) and not pd.isna(n2):
                    df.loc[i, 'PulseRate'] = (n1 + n2) / 2
                    continue

            # 4. Trung bình tất cả huyết áp của người đó
            mean_person = group['PulseRate'].mean()
            if not pd.isna(mean_person):
                df.loc[i, 'PulseRate'] = mean_person
                continue

            # 5. Trung bình theo giới tính
            mean_sex = df[df['Sex'] == df.loc[i, 'Sex']]['PulseRate'].mean()
            if not pd.isna(mean_sex):
                df.loc[i, 'PulseRate'] = mean_sex
                continue

            # 6. Trung bình toàn bộ dữ liệu
            df.loc[i, 'PulseRate'] = df['PulseRate'].mean()

print(df)
```

```python
# Câu 12: Rút gọn và reindex, lưu file
df = df[['Id', 'Age', 'Weight', 'Firstname', 'Lastname', 'PulseRate', 'Sex', 'Time']]
df = df.reset_index(drop=True)
df.to_csv('patient_heart_rate_clean.csv', index=False)
print(df)
print("Đã lưu file patient_heart_rate_clean.csv")
```
